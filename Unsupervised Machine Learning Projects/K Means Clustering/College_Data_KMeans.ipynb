```python
# Create and train model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(college_df.drop('private',axis=1))
```

```python
# Cluster center vector
kmeans.cluster_centers_
```

```python
# Function to convert 'private' column to what machine can read
def converter(private):
    if private == 'Yes':
        return 1
    else:
        return 0    
```

```python
# Creating a new column 'Cluster'
college_df['Cluster'] = college_df['private'].apply(converter)
```

```python
# Head and tail of the data showing the newly added 'Cluster' column
college_df
```

```python
# Checking the confusion matrix and classification report of the kmeans cluster
from sklearn.metrics import confusion_matrix, classification_report

# Confusion Matrix
print(confusion_matrix(college_df['Cluster'], kmeans.labels_))
print('\n')

# Classification Report
print(classification_report(college_df['Cluster'], kmeans.labels_))
```
