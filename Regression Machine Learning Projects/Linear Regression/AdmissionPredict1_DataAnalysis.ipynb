```python
# Split columns into X and y
X = Adpred1_df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']]
y = Adpred1_df['Chance of Admit ']
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# Create and train model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Coefficient of the data
print('Coefficient: \n\n',model.coef_)
```

```python
# Prediction
pred = model.predict(X_test)
```

```python
# Scatter Plot of the y_test(real test value) and pred1(predicted value)
plt.scatter(y_test, pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Regression Metric Library and Variance Score
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score

# Mean Absolute Error
print('MAE: ',mean_absolute_error(y_test, pred))

# Mean Squared Error
print('MSE: ',mean_squared_error(y_test, pred))

# Root Mean Squared Error
print('RMSE: ',np.sqrt(mean_squared_error(y_test, pred)))
print('\n')

# Explained Variance Score
print('Variance Score: ',explained_variance_score(y_test,pred))
```

```python
# Residual Dsiplot
sns.distplot((y_test-pred), bins=50)
```

```python
# CDF
cdf = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
cdf
```
