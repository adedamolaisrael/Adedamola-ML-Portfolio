```python
# Split the columns into X and y
X = housing_df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = housing_df['Price']
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Create and train model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Coefficient of the model
print('Coefficient = ',model.coef_)
```

```python
# Prediction of the model
predictions = model.predict(X_test)
```

```python
# Scatterplot to compare the predicted_y and tested_y
plt.scatter(y_test, predictions)
plt.xlabel('Tested Y')
plt.ylabel('Predicted y')
```

```python
# Regression Metric Library
from sklearn import metrics

# Mean Absolute Error
print('MAE = ',metrics.mean_absolute_error(y_test, predictions))

# Mean Squared Error
print('MSE = ',metrics.mean_squared_error(y_test, predictions))

# Root Mean Squared Error
print('RMSE = ',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Variance Score
print('Variance Score = ',metrics.explained_variance_score(y_test, predictions))
```

```python
# Residual Distplot
sns.distplot(y_test-predictions, bins=50)
```

```python
# CDF
cdf = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
cdf
```
