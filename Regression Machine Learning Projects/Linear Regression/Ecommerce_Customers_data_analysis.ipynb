```python
# Spliting the data into training and test set
X = customers_df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = customers_df['Yearly Amount Spent']
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Creating and training the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Printing the coefficients of the model
print('Coefficients: \n',model.coef_)
```

```python
# Prediction
predictions = model.predict(X_test)
```

```python
# Scatterplot of real test values and predicted values
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Loading regression metrics library
from sklearn import metrics

# Mean Absolute Error
print('MAE = ', metrics.mean_absolute_error(y_test, predictions))

# Mean Squared Error
print('MSE = ', metrics.mean_squared_error(y_test, predictions))

# Root Mean Squared Error
print('RMSE = ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Explain variance score
print('Variance Score = ', metrics.explained_variance_score(y_test, predictions))
```

```python
# Exploring the residual
sns.distplot((y_test-predictions), bins=50)
```

```python
# How to boost the yearly amount spent by the customers
cdf = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
cdf
```
