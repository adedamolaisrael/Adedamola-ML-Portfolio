```python
# Import all the needed libraries
import numpy as np
import pandas as pd
```

```python
# Turn off all warnings
import warnings
warnings.filterwarnings('ignore')
pd.options.mode.chained_assignment=None
```

```python
# Import the visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read CSV(dataset)
from sklearn.datasets import load_boston
HousePrice_dataset = load_boston()
```

```python
print(HousePrice_dataset)
```

```python
# Loading the data into Pandas DataFrame
HousePrice_df = pd.DataFrame(data=HousePrice_dataset.data, columns=HousePrice_dataset.feature_names)
```

```python
# Taking a look at the dataframe
HousePrice_df
```

```python
# Adding of the target column (PRICE) to the dataframe
HousePrice_df['PRICE'] = HousePrice_dataset.target
```

```python
# Taking a look at the dataframe again
HousePrice_df
```

```python
# Shape of the dataframe
print('Number of rows: ',HousePrice_df.shape[0])
print('Number of columns: ',HousePrice_df.shape[1])
```

```python
# Info of the dataframe
HousePrice_df.info()
```

```python
# Statistical measures 
HousePrice_df.describe()
```

```python
# Check for missing values
HousePrice_df.isnull().sum()
```

# Exploratory Data Analysis

```python
# Setting up the visualization environment
sns.set_palette('GnBu_r')
sns.set_style('whitegrid')
```

```python
# PairPlot
sns.pairplot(HousePrice_df)
```

```python
# HeatMap
plt.figure(figsize=(10,10))
sns.heatmap(HousePrice_df.corr(), annot=True, cmap='coolwarm', cbar=True, square=True, annot_kws={'size':8})
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(HousePrice_df.corr(), annot=True, cmap='coolwarm')
```

```python
# Split data to X and y
X = HousePrice_df.drop(['PRICE'], axis=1)
y = HousePrice_df['PRICE']
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Create and Train model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Coefficient of the data
print('Coefficient: \n',model.coef_)
```

```python
# Prediction
pred = model.predict(X_test)
```

```python
# ScatterPlot of the y_test(real test value) and pred(predicted value)
plt.scatter(y_test, pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.title('Y Test Vs Predicted Y')
```

```python
# Evaluating with regression metrics libraries
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score

# Mean Absolute Error
print('MAE = ', mean_absolute_error(y_test, pred))

# Mean Squared Error
print('MSE = ', mean_squared_error(y_test, pred))

# Root Mean Squared Error
print('RMSE = ', np.sqrt(mean_squared_error(y_test, pred)))

print('\n')

# Explain variance score
print('Variance Score = ', explained_variance_score(y_test, pred))
```

```python
# Residual Distplot
sns.distplot((y_test-pred), bins=50)
```

```python
# CDF
cdf = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
cdf
```
