```python
# load libraries
import pandas as pd
```

```python
# load dataset
# from pandas import read_csv
filename = 'iris.csv'
dataframe = pd.read_csv(filename, encoding='utf-8')
```

```python
# Descriptive Statistics
# Shape
print("Number of rows::", dataframe.shape[0])
print("Number of columns::", dataframe.shape[1])
```

```python
# Tail
print(dataframe.tail())
```

```python
import seaborn as sns
%matplotlib inline
```

```python
# Distribution Plot
sns.distplot(dataframe['petal_length'])
```

```python
# Distribution Plot
sns.distplot(dataframe['petal_width'])
```

```python
# Distribution Plot
sns.distplot(dataframe['sepal_length'])
```

```python
# Distribution Plot
sns.distplot(dataframe['sepal_width'])
```

```python
# Joint Plot
sns.jointplot(x='petal_length', y='petal_width', data=dataframe)
```

```python
# Joint Plot
sns.jointplot(x='sepal_length', y='sepal_width', data=dataframe)
```

```python
# Joint Plot
sns.jointplot("sepal_length", "sepal_width", data=dataframe, color="k").plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
# Joint Plot
sns.jointplot("sepal_length", "sepal_width", data=dataframe, marginal_kws=dict(bins=15, rug=True), annot_kws=dict(stat="r"), s=40, edgecolor="w", linewidth=1)
```

```python
# Pair Plot
sns.pairplot(dataframe)
```

```python
# Pair Plot
sns.pairplot(dataframe,hue='species')
```

```python
# Rug Plot
sns.rugplot(dataframe['petal_length'])
```

```python
# Bar Plot
sns.barplot(x='species', y='petal_length', data=dataframe)
```

```python
# Bar Plot
sns.barplot(x='species', y='petal_width', data=dataframe)
```

```python
# Bar Plot
sns.barplot(x='species', y='sepal_length', data=dataframe)
```

```python
# Bar Plot
sns.barplot(x='species', y='sepal_width', data=dataframe)
```

```python
# Count Plot
sns.countplot(x='species', data=dataframe)
```

```python
# Box and Whisker Plot
sns.boxplot(x='species', y='petal_width', data=dataframe)
```

```python
# Box and Whisker Plot
sns.boxplot(x='species', y='petal_length', data=dataframe)
```

```python
# Box and Whisker Plot
sns.boxplot(x='species', y='sepal_width', data=dataframe)
```

```python
# Box and Whisker Plot
sns.set(style="whitegrid")
sns.boxplot(x='species', y='sepal_width', data=dataframe, hue='species')
```

```python
# Violin Plot
sns.violinplot(x='species', y='sepal_length', data=dataframe)
```

```python
# Strip Plot
sns.stripplot(x='species', y='sepal_length', data=dataframe, jitter=0.05)
```

```python
# Swarm Plot
sns.swarmplot(x='species', y='sepal_length', data=dataframe)
```

```python
# Swarm Plot mixed with violin Plot
sns.violinplot(x='species', y='sepal_length', data=dataframe)
sns.swarmplot(x='species', y='sepal_length', data=dataframe, color='black')
```

```python
# Factor Plot
sns.catplot(x='species', y='sepal_length', data=dataframe)
```

```python
# Heat Map
tc = dataframe.corr()
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
# Lm plot
sns.lmplot(x='sepal_length', y='sepal_width', data=dataframe, hue='species')
```

```python
# PairGrid
import matplotlib.pyplot as plt
g = sns.PairGrid(dataframe)
g.map(plt.scatter)
```

```python
# Facet Grid
h = sns.FacetGrid(dataframe, col='species')
h.map(sns.distplot, 'sepal_length')
```
